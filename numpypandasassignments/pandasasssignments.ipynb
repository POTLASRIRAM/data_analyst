```python
### Assignment 4: Data Aggregation

#1. Create a Pandas DataFrame with 2 columns: 'Category' and 'Value'.
#  Fill the 'Category' column with random categories ('A', 'B', 'C') and the 'Value' column with random integers.
#  Group the DataFrame by 'Category' and compute the sum and mean of 'Value' for each category. 
s1 = pd.Series(np.random.choice(['A','B','C'],size=20))
s2 = pd.Series(np.random.randint(1,10,size=20))

df = pd.concat([s1,s2],axis=1)
df.columns = ['Category','Value']
grouped_mean = df.groupby(['Category'])['Value'].mean()
grouped_sum = df.groupby(['Category'])['Value'].sum()

print("Grouped sum",grouped_sum)
print("Grouped mean",grouped_mean)
```

```python
#2. Create a Pandas DataFrame with 3 columns: 'Product', 'Category', and 'Sales'. 
# Fill the DataFrame with random data. Group the DataFrame by 'Category' and compute the total sales for each category. 
category = pd.Series(np.random.choice(['A','B','C'],size=20))
sales = pd.Series(np.random.randint(1,5,size=20))
product = pd.Series(np.random.choice(['TV','AC','Laptop','Computer'],size=20))

df = pd.concat([product,category,sales],axis=1)
df.columns = ['Product','Category','Sales']
df

```

```python
grouped_sales = df.groupby(['Category'])['Sales'].sum()
print(grouped_sales)
```
