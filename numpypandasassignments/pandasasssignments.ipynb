```python
### Assignment 4: Data Aggregation

#1. Create a Pandas DataFrame with 2 columns: 'Category' and 'Value'.
#  Fill the 'Category' column with random categories ('A', 'B', 'C') and the 'Value' column with random integers.
#  Group the DataFrame by 'Category' and compute the sum and mean of 'Value' for each category. 
s1 = pd.Series(np.random.choice(['A','B','C'],size=20))
s2 = pd.Series(np.random.randint(1,10,size=20))

df = pd.concat([s1,s2],axis=1)
df.columns = ['Category','Value']
grouped_mean = df.groupby(['Category'])['Value'].mean()
grouped_sum = df.groupby(['Category'])['Value'].sum()

print("Grouped sum",grouped_sum)
print("Grouped mean",grouped_mean)
```

```python
### Assignment 6: Time Series Analysis

#1. Create a Pandas DataFrame with a datetime index and one column filled with random integers. 
# Resample the DataFrame to compute the monthly mean of the values.
# resample groups all the monthly wise dates and calcukates the mean of all numeric values in that date.
dates=pd.date_range(start='2025-01-01',end='2025-03-01',freq='D')
df = pd.DataFrame(dates,columns=['Date'])
df['Values'] = np.random.randint(1,10,size=len(dates))
df.set_index('Date',inplace=True)
df
```

```python
monthly_mean=df.resample('M').mean()
monthly_mean
```

 big difference between normal mean and rolling mean in pandas:

 1. Normal Mean (df.mean() or df.resample().mean())
What it does: Calculates the average across the whole column or within fixed resample groups (like by month).

Use case: Gives an overall summary statistic.

Example: df['data'].mean() gives the average of all values in the 'data' column.

 2. Rolling Mean (df.rolling(window).mean())
What it does: Calculates the average over a sliding window across the data.

Use case: Useful for trend analysis, smoothing data, or detecting changes over time.

Example:
df['data'].rolling(window=7).mean()
 gives a 7-day moving average of the 'data' column.

```python
#2.Create a Pandas DataFrame with a datetime index ranging from '2021-01-01' to '2021-12-31'
#  and one column filled with random integers. Compute the rolling mean with a window of 7 days.
dates =pd.date_range(start = '2021-01-01' ,end= '2021-12-31',freq='D')
df = pd.DataFrame(dates,columns = ['Date'])
df['Value'] = np.random.randint(1,20,size=len(dates))
df.set_index('Date',inplace=True)
rolling_mean = df.rolling(window=7).mean()
rolling_mean
```

```python
### Assignment 8: Pivot Tables

#1. Create a Pandas DataFrame with columns 'Date', 'Category', and 'Value'.
#  Create a pivot table to compute the sum of 'Value' for each 'Category' by 'Date'.
#A Pandas pivot table is a powerful tool to summarize, aggregate, and rearrange data  similar to Excel pivot tables.
#pd.pivot_table(data=df,index='which colmn we want to take as index',columns='text columns',values='numerical cols',aggfunc='sum,mean,median,std,var')
dates = pd.date_range(start='2025-01-01',end='2025-02-01',freq='D')
df = pd.DataFrame(dates,columns=['Date'])
df['Category'] = np.random.choice(['A','B','C'],size=len(dates))
df['Value'] = pd.Series(np.random.randint(1,10,size=len(dates)))
pivot_table= pd.pivot_table(data=df,index='Date',columns='Category',values='Value',aggfunc='sum')
pivot_table
```

```python
#2. Create a Pandas DataFrame with columns 'Year', 'Quarter', and 'Revenue'. 
# Create a pivot table to compute the mean 'Revenue' for each 'Quarter' by 'Year'.
dates = pd.date_range(start='2021',end='2022',freq='W')
df = pd.DataFrame(np.random.choice(['2020','2021','2022','2024'],size=(20)),columns=['Date'])
df['Quarter'] = np.random.choice(['Q1','Q2','Q3','Q4'],size=len(df['Date']))
df['Revenue'] = pd.Series(np.random.randint(10000,100000,size=len(df['Date'])))
df
```

```python
pivot_table = pd.pivot_table(data=df,index='Date',columns='Quarter',values='Revenue',aggfunc='mean')
pivot_table
```
