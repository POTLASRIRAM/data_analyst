```python
import pandas as pd
```

```python
## Series -- A Pandas Series is a one-dimensional array-like object that can hold any data type. It is similar to a column in a table.

ls = [1,2,3,4,5,6,7]
series = pd.Series(ls)
series.name = "Values"
print(series)
```

```python
## Creating series from a directory

d = {"a":1,"b":2,"c":3,"d":4,"e":5,"f":6,"g":7}
dict_series = pd.Series(d)
dict_series.name = "Dict_Pandas_Series"
print(dict_series)
```

```python
## Can't create a series from set as it is un ordered
dd = {1,2,3,4,5,6,7}
print(type(dd))
index = ['c','j','e','f','g','h','i']
ext_index = pd.Series(dd,index = index)
print(ext_index)
```

```python
## Can create a series from tuple as it is ordered
dd = (1,2,3,4,5,6,7)
print(type(dd))
index = ['c','j','e','f','g','h','i']
ext_index = pd.Series(dd,index = index)
print(ext_index)
```

## Data Frames

```python
## Data Frames -- A two dimentional matrices
## Creating data frame from dictionary of list

dict_lst = {
    'Name' : ['a','b','c'],
    'Age': [21,22,23],
    'City' :['NYC','HYD','BAN']
}

df = pd.DataFrame(dict_lst)
df
```

```python
## Adding two series into a data frame
names = ['a','b','c']
age = [56,67,89]

lst1 = pd.Series(names,name = "names")
lst2 = pd.Series(age,name = "age")
df1 = pd.concat([lst1,lst2],axis =1)
df1
```

```python
## Create a Data frame From a List of Dictionaries

data=[
    {'Name':'Krish','Age':32,'City':'Bangalore'},
    {'Name':'John','Age':34,'City':'Bangalore'},
    {'Name':'Bappy','Age':32,'City':'Bangalore'},
    {'Name':'JAck','Age':32,'City':'Bangalore'}
    
]
df=pd.DataFrame(data)
df
```

###### Locating particular data using the loc and iloc properties
###### loc :	Label-based selection	Access data by row/column names
###### iloc : Index-based selection	Access data by row/column numbers (positions)

```python
df2.loc[0]
```

```python
df2.iloc[0,0]
```

```python
## Accessing column using iloc
## If you want multiple columns use list of columns
df2.loc[0:4,['Category']]
```

```python
df2.loc[0:4,['Category','Value']]
```

```python
df2.iloc[0:4,1:4]
```

```python
df2.iloc[0,0]
```

```python
## Accessing a single cell using 'at' 
# at -- Access one single value by label and this is Fastest EX :	df.at['row', 'col']
# iat -- Access one single value by position and this is Fastest EX :	df.iat[0,1]

df2.at[0,'Category']
```

```python
df2.iat[0,1]
```
